```python
import pandas as pd
from functools import reduce

"""
Needs testing and complete methods revision BT 2021-12-3

Puzzle 2: Obtain the oxygen and CO2 scrubber rating
oxygen rating - binary num remaining after sequentially 
collecting most common bits in each input pos

CO2 scrubber rating - binary num remaining after sequentially 
collecting least common bits in each input pos

If bits are equal, 1 assigned to oxygen, 0 assigned to C02

Method: csv_to_bitpos_df()->df_to_bincoords()->coord_mul()
1) input -> df with each digit in its own col
2) oxygen = [], CO2 = []
3) enumerate for loop through each col
4) filter by most/least for oxygen/CO2 from prev bitsums, if applicable
5) sum col -> 1's
6) subtract sum from num of rows -> 0's in each pos
7) determine >, <, or = for bits -> append current bitsum
8) oxygen, CO2 int list -> str list for concat
9) oxygen_concat.join(), CO2_concat.join()
10) oxygen_concat, CO2_concat -> (int(oxygen_concat, 2), int(CO2_concat))
11) Multiply rates -> print resulting life support rating to stdout
"""

def csv_to_bitpos_df(path: str) -> pd.DataFrame:
    with open(path, 'r') as file:
        lines = [list(map(int, line.rstrip('\n'))) for line in file if line.strip()] # remove \n and convert each line to list of digits
    return(pd.DataFrame(lines)) # 1)

def rating_filter(df: pd.DataFrame, rating_type: str)->int:
    rating = []
    rating_df = df.copy()
    for colname in df:
        if colname != 0:
            rating_df = rating_df[rating_df[colname-1]==rating[colname-1]]
            if len(rating_df) == 1:
                rating = rating_df.iloc[0,:].to_list()
                break
        rating_1bit = rating_df[colname].sum()
        if rating_type == 'o':
            if rating_1bit >= len(rating_df)/2:
                rating.append(1)
            else:
                rating.append(0)
        else:
            if rating_1bit >= len(rating_df)/2:
                rating.append(0)
            else:
                rating.append(1)
    rating_str = [str(r) for r in rating]
    rating_concat = "".join(rating_str)
    return(int(rating_concat, 2))

def coord_mul(incoords: Tuple[int, int]) -> int:
    return(reduce(lambda x, y: x*y, incoords)) # 11) reduce to apply lambda mul to all eles

def main() -> int:
    input_df = csv_to_bitpos_df('Day-3-input.txt')
    oxygen = rating_filter(input_df, 'o')
    CO2 = rating_filter(input_df, 'c')
    print(coord_mul((oxygen, CO2)))

main()
```
